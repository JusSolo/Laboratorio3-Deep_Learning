```python
# imports
import kagglehub
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from pathlib import Path
import numpy as np # linear algebra
import pandas as pd #
from tensorflow.keras.utils import to_categorical
```

```python
import kagglehub

# Download latest version
dataset_path = kagglehub.dataset_download("agungpambudi/mnist-multiple-dataset-comprehensive-analysis")

print("Path to dataset files:", dataset_path)
```

```python
def load_images_and_labels_from_filenames(base_path):
    base_path = Path(base_path)
    images = []
    labels = []

    # Buscar imágenes en cualquier subdirectorio
    for img_path in base_path.rglob("*.png"):  # cambia a *.jpg si es necesario
        filename = img_path.name
        try:
            label = filename.split('.')[1]  # Toma el número después del primer punto
            img = cv2.imread(str(img_path), cv2.IMREAD_GRAYSCALE)
            if img is not None:
                img = cv2.resize(img, (28, 28))  # Tamaño estándar
                images.append(img)
                labels.append(label)
        except IndexError:
            print(f"⚠️ No se pudo extraer etiqueta de: {filename}")
    
    return np.array(images), np.array(labels)
```

```python
images, labels = load_images_and_labels_from_filenames(dataset_path)
images = images.astype("float32") / 255.0
```

```python
import matplotlib.pyplot as plt

def plot_first_images(images, labels, n=10):
    plt.figure(figsize=(12, 2))
    for i in range(n):
        plt.subplot(1, n, i + 1)
        plt.imshow(images[i])
        plt.title(f"Etiqueta: {labels[i]}")
        plt.axis("off")
    plt.tight_layout()
    plt.show()

# Ejecutar la función
plot_first_images(images, labels, n=10)
```

```python
labels = to_categorical(labels, 10)
```

```python
from sklearn.model_selection import train_test_split

# 1. Separar en temporal y test (80% temp, 20% test)
X_temp, X_test, y_temp, y_test = train_test_split(
    images, labels, test_size=0.2, random_state=42, stratify=labels
)

# 2. Separar temp en train y val (75% train, 25% val => 60% y 20% del total)
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.25, random_state=42, stratify=y_temp
)

# 3. Mostrar tamaños
print(f"Train: {len(X_train)} imágenes")
print(f"Validation: {len(X_val)} imágenes")
print(f"Test: {len(X_test)} imágenes")
print(images[1].shape)
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import classification_report
import numpy as np
import pandas as pd
import keras_tuner as kt
import os
labels[0]
```

```python

def model_builder(input_shape=(28, 28, 1), num_classes=10):
    model = tf.keras.Sequential()
    model.add(tf.keras.layers.Input(shape=input_shape))
    
    
    # Bloque 1 
    # capa convolucional 64 filtros
    model.add(tf.keras.layers.Conv2D(filters=64, kernel_size=(3, 3), 
                                    activation='relu', padding='same'))
    # convolucional 64 filtros
    model.add(tf.keras.layers.Conv2D(filters=64, kernel_size=(3, 3), 
                                    activation='relu', padding='same'))
    # Max Pooling 2*2
    model.add(tf.keras.layers.MaxPooling2D(pool_size=(2, 2)))
    
    # Bloque 2 
    # capa convolucional 128 filtros
    model.add(tf.keras.layers.Conv2D(filters=128, kernel_size=(3, 3), 
                                    activation='relu', padding='same'))
    # capa convolucional 128 filtros
    model.add(tf.keras.layers.Conv2D(filters=128, kernel_size=(3, 3), 
                                    activation='relu', padding='same'))
    # Max Pooling 2x2 
    #model.add(tf.keras.layers.MaxPooling2D(pool_size=(2, 2)))

    
    # espaquetizacion del tensor 
    model.add(tf.keras.layers.Flatten())
    
    # Capas Densas 
    model.add(tf.keras.layers.Dense(256, activation='relu'))
    model.add(tf.keras.layers.Dense(64, activation='relu'))
    model.add(tf.keras.layers.Dense(16, activation='relu'))
    
    # Capa de salida
    model.add(tf.keras.layers.Dense(num_classes, activation='softmax'))
    
    model.compile(optimizer='adam',
                 loss='categorical_crossentropy',
                 metrics=['accuracy'])
    
    return model
```

```python
model = model_builder(input_shape=(28, 28, 1), num_classes=10)

history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=5,
    verbose=1
)


# Graficar accuracy
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy durante el entrenamiento')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Predicciones (probabilidades)
y_pred_probs = model.predict(X_test)

# Convertimos a clases predichas (índice de la probabilidad más alta)
y_pred = np.argmax(y_pred_probs, axis=1)

```

```python

y_test_classes = np.argmax(y_test, axis=1)
class_names = [str(i) for i in range(10)]
# Calcular matriz de confusión
cm = confusion_matrix(y_test_classes, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
disp.plot(cmap='Blues', xticks_rotation=45)
plt.title("Matriz de confusión")
plt.show()
```
